```python
import requests as req
import pandas as pd 

res = req.get('https://jsonplaceholder.typicode.com/posts')
if res.status_code == 200:
    posts = res.json()
    df = pd.DataFrame(posts)
    print(df.head())
else:
    print(f"Failed to retrieve data: {res.status_code}")
```

```python
import requests as req
from bs4 import BeautifulSoup

url = 'http://quotes.toscrape.com'
res = req.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
quotes = soup.find_all('span', class_='text')
authors = soup.find_all('small', class_='author')
for quote, author in zip(quotes, authors):
    print(f"{quote.text} - {author.text}")
```

```python
import pandas as pd
quote_list = []

for quote, author in zip(quotes, authors):
    quote_list.append({'quote': quote.text, 'author': author.text})

df = pd.DataFrame(quote_list)
print(df.head())
```

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('students.db')

data = {
    'student_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'major': ['Math', 'Physics', 'Chemistry', 'Biology', 'Computer Science'],
    'gpa': [3.5, 3.7, 3.2, 3.8, 3.9]
}

df = pd.DataFrame(data)
df.to_sql('students', conn, if_exists='replace', index=False)
query = "SELECT * FROM students WHERE gpa > 3.7"
result = pd.read_sql_query(query, conn)
print(result)
conn.close()
```
